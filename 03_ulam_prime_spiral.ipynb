```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import animation
from IPython.display import HTML, display
import warnings; warnings.filterwarnings('ignore')

plt.rcParams.update({
    'figure.facecolor':'#03020a','axes.facecolor':'#03020a',
    'axes.edgecolor':'#1a0a30','text.color':'#c8b8ff',
    'axes.labelcolor':'#c8b8ff','xtick.color':'#5030a0',
    'ytick.color':'#5030a0','font.family':'monospace'
})
```

```python
def sieve(n):
    p=np.ones(n+1,dtype=bool); p[0]=p[1]=False
    for i in range(2,int(n**.5)+1):
        if p[i]: p[i*i::i]=False
    return p

def build_ulam(side=601):
    N=side*side; primes=sieve(N+4); grid=np.zeros((side,side),np.float32)
    x=y=side//2; n=1; step=1; count=0; di=0; turns=0
    dirs=[(0,1),(-1,0),(0,-1),(1,0)]
    for _ in range(N):
        if 0<=y<side and 0<=x<side and primes[n]: grid[y,x]=1.
        dx,dy=dirs[di%4]; x+=dx; y+=dy; n+=1; count+=1
        if count==step:
            count=0; di+=1; turns+=1
            if turns%2==0: step+=1
    return grid

print("Building Ulam spiral ...", end=" ")
grid=build_ulam(601); pc=int(grid.sum())
print(f"done. {pc:,} primes ({100*pc/601**2:.1f}% of range)")

cu=LinearSegmentedColormap.from_list("ul",["#03020a","#1a0a00","#804000","#ffbe0b","#fff7cc"])
fig,axes=plt.subplots(1,3,figsize=(21,7),facecolor='#03020a')
fig.subplots_adjust(left=.02,right=.98,top=.88,bottom=.04,wspace=.04)

axes[0].imshow(grid,cmap=cu,origin='lower',interpolation='nearest')
axes[0].set_title("Full spiral  1 → 361,201\nLit pixel = prime",color='#ffbe0b',fontsize=9,pad=5); axes[0].axis('off')

c=601//2; z=80
axes[1].imshow(grid[c-z:c+z,c-z:c+z],cmap=cu,origin='lower',interpolation='nearest')
axes[1].set_title("Centre zoom  160×160\nDiagonals = prime-rich quadratics",color='#ffbe0b',fontsize=9,pad=5); axes[1].axis('off')

from numpy.fft import fft2,fftshift
spec=np.abs(fftshift(fft2(grid-grid.mean())))**.3
cf=LinearSegmentedColormap.from_list("ft",["#03020a","#002a40","#00aaff","#fff"])
axes[2].imshow(spec,cmap=cf,origin='lower',interpolation='bilinear')
axes[2].set_title("2D Fourier power spectrum\nOff-axis spikes ≠ random",color='#ffbe0b',fontsize=9,pad=5); axes[2].axis('off')

fig.suptitle("ULAM PRIME SPIRAL  ·  Primes self-organise along quadratic diagonals  ·  Reason: unproved",
             color='#ffbe0b',fontsize=12,y=.97)
plt.show()
print("Primes are not random. The Riemann Hypothesis would partially explain why.")
```
